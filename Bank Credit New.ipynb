## 2.9 Encoding categorical variables

## Applying Neural Network

```python
import tensorflow as tf
import keras
```

```python
# applying neural network
bcc_x = np.array(bcc_x)
bcc_y = np.array(bcc_y)
```

```python
from sklearn.model_selection import train_test_split
bcc_x_train, bcc_x_test, bcc_y_train, bcc_y_test = train_test_split(bcc_x, bcc_y, test_size = 0.2, random_state = 555)
```

```python
bcc_x_train = np.array(bcc_x_train)
bcc_x_test = np.array(bcc_x_test)
bcc_y_train = np.array(bcc_y_train)
```

```python
for i in range(len(bcc_x_train)):
    bcc_x_train[i,]
```

```python
for i in range(len(bcc_x_test)):
    bcc_x_test[i,]
```

```python
# normalize the data
bcc_x_train = tf.keras.utils.normalize(bcc_x_train)
bcc_x_test = tf.keras.utils.normalize(bcc_x_test)
```

```python
# build the neural network
model = tf.keras.models.Sequential()    # initilize the netwaork

model.add(tf.keras.layers.Dense(256, activation = tf.nn.relu))  # 1st hidden layer with 256 neuron & activation is relu
model.add(tf.keras.layers.Dense(256, activation = tf.nn.relu))  # 2nd hidden layer with 256 neuron & activation is relu
model.add(tf.keras.layers.Dense(256, activation = tf.nn.relu))  # 3rd hidden layer with 256 neuron & activation is relu
model.add(tf.keras.layers.Dense(10, activation = tf.nn.softmax))  # for o/p layer, use softmax instead of relu

adam = tf.keras.optimizers.Adam(lr = 0.001)

model.compile(optimizer = adam, loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(bcc_x_train, bcc_y_train, batch_size = 100, epochs = 200, validation_split = 0.2)
```

```python
model.predict_classes(bcc_x_test)
```

```python
tabnn = confusion_matrix(bcc_y_test, model.predict_classes(bcc_x_test))
tabnn
```

```python
accuracy = accuracy_score(model.predict_classes(bcc_x_test), bcc_y_test)
accuracy
```
